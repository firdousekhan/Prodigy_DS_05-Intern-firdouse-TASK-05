```python
db2=db[db['State']=='CA']
```

```python
db2['IDD'] = db2['ID'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
```

```python
db2
```

```python
db2=db2.dropna(subset=['Temperature(F)','Wind_Chill(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Direction', 'Wind_Speed(mph)',
                      'Weather_Condition','City','Sunrise_Sunset',
       'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight'])
```

```python
db_num=db2.select_dtypes(np.number)
col_name=[]
length=[]

for i in db_num.columns:
    col_name.append(i)
    length.append(len(db_num[i].unique()))

db3 = pd.DataFrame(zip(col_name,length),columns=['feature','count_of_unique_values'])
db3
```

```python
plt.figure(figsize=(15 ,9))
sns.heatmap(db_num.corr() , annot=True)
```

```python
accidents_by_cities = db2['City'].value_counts()
accidents_by_cities
```

```python
fig, ax = plt.subplots(figsize=(8,5))
accidents_by_cities[:10].plot(kind='bar')
ax.set(title = 'Top 10 cities By Number of Accidents',
       xlabel = 'Cities',
       ylabel = 'Accidents Count')
plt.show()
```

```python
weather_conditions = db2['Weather_Condition'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
weather_conditions.sort_values(ascending=False)[:20].plot(kind='bar', ax=ax)
ax.set(title='Weather Conditions at Time of Accident Occurrence',
       xlabel='Weather',
       ylabel='Accidents Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
